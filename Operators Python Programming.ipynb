***ARITHMETIC OPERATOR.***

***Add (+)***

```python
# sum (+) the two numbers.
x = 7  
y = 6
z = x + y
print(z)
```

***Subtract (-)***

```python
# sub (-) the two numbers.
x = 7
y = 6
z = x - y 
print(z)
```

***Multyply (*)***

```python
# multyply (*) two numbers.
x = 7
y = 6
z = x * y
print(z)
```

***Division (/)***

```python
# Division (/) the two numbers.
x = 42
y = 6
z = x / y
print(z)
```

***Floor Division (//)***

```python
# Floor Division (//) the two numbers.
x = 7
y = 6
z = x // y
print(z)
```

```python
# Modulas Division (%) the two numbers.
x = 8
y = 6
z = x % y
print(z)
```
