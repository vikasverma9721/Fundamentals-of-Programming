## Assignment Operator

```python
# Add And Assignment
x=10
x+=2
print(x)
```

```python
# Subtract And Assignment
x=10
x-=2
print(x)
```

```python
# Multiply And Assignmen
x=10
x*=2
print(x)
```

```python
# Division And Assignment
x=10
x/=2
print(x)
```

```python
# Fllor Division And Assignment
x=10
x//=2
print(x)
```

```python
# Moduls Division And Assignment
x=10
x%=2
print(x)
```
